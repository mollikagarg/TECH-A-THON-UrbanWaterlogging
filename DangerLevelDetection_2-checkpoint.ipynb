```python
## IMPORTS

# used for reading images from data 
import cv2

# used for manipulating directory paths
import os

# Scientific and vector computation for python
import numpy as np

# For image visualisation
import matplotlib.pyplot as plt

# used for reading excel file
import pandas as pd

# used for saving the model
import pickle

# used for getting features from images
from skimage import feature as ft

# used for training SVM
from sklearn.svm import SVC

# used for testing model's accuracy score
from sklearn.metrics import accuracy_score

# used for training random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
## DATA DIRECTORY PATHS 

datapath="C:\\Users\\molli\\Downloads\\data\\data"
floodTrainPath=os.path.join(datapath,'train','flooding')
floodValPath=os.path.join(datapath,'valid','flooding')
floodTestPath=os.path.join(datapath,'test','flooding')
```

```python
## FUNCTION TO READ DATA 

def read_data(floodpath):
    """
    Loading RGB images of a given split from flood and normal image directory
    
    Parameters
    ----------
    floodpath: string
               it represents the path of the flood folder images directory
    
    Returns
    -------
    floodImg: list
              it contains all the images present in flood folder
              
    floodImgPath: list
              it contains all the paths of images present in flood folder
    """
    floodImgPath=os.listdir(floodpath)
    floodImg=[]
    for i in range(len(floodImgPath)):
        currentPath=os.path.join(floodpath,floodImgPath[i])
        img=cv2.imread(currentPath)
        floodImg.append(img)
        
    return floodImg, floodImgPath
```

### Different Features Extraction Methods

```python
## 1) Histogram of gradient feature method(hog feature)

def hog_feature(img_array, resize=(256,192)):
    """
    Extract hog feature from an image
    
    Parameters
    ----------
    img_array: ndarray
               it is an image array
               
    resize: tuple
            it represents size of the image for extracture
    
    
    Returns
    -------
    features: ndarray 
              it is an image feature vector     
    """
    img = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
    img = cv2.resize(img, resize)
    bins = 9
    cell_size = (8, 8)
    cpb = (2, 2)
    norm = "L2"
    features = ft.hog(img, orientations=bins, pixels_per_cell=cell_size, 
                        cells_per_block=cpb, block_norm=norm, transform_sqrt=True)
    return features
```

```python
## 2) RGB Features for images

def rgb_feature(img_array, resize=(128,96)):
    """
    Extract RGB feature from an image
    
    Parameters
    ----------
    img_array: ndarray
               it is an image array
               
    resize: tuple
            it represents size of the image for extracture
    
    
    Returns
    -------
    features: ndarray 
              it is an image feature vector       
    """
    img = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
    img = cv2.resize(img, resize)
    img= img/255.0
    features=img.reshape(-1)
   
    return features
```

### Data Augmentation

```python
#flip image left to right
def flip_img(img):
    return np.fliplr(img)
```

### Reading Data

```python
#read training data

#reading training annotations from excel file
TrainDangerYlabel=pd.read_excel("C:\\Users\\molli\\OneDrive\\Desktop\\TrainDangerLevel.xlsx")

floodImg, floodImgPath = read_data(floodTrainPath)

# calculate hog features for each image
# y_labels is 0 for low water logging
# y_lablels is 1 for high water logging 


# hogFeatures(dimension: m x n) contain train features
# y_labels(dimensin: m x 1) contain training labels

hogFeature=[]
y_labels=[]
for i in range(len(floodImg)):
    k1=hog_feature(floodImg[i])
    hogFeature.append(k1)
    currentName= floodImgPath[i]
    imgno=currentName.split("_")[1]
    imgno=int(imgno.split(".")[0])
    y=TrainDangerYlabel.loc[TrainDangerYlabel["Image No."]==imgno]["New Level"].item()
    y_labels.append(y)
    
    #data augmentation
    flipimg = flip_img(floodImg[i])
    k=hog_feature(flipimg)
    hogFeature.append(k)
    y_labels.append(y)
    
# shuffling the training data   
rand_col=np.arange(0,len(y_labels),dtype=np.int)
np.random.shuffle(rand_col)
y_labels=np.array(y_labels)[rand_col.astype(int)]
hogFeature=np.array(hogFeature)[rand_col.astype(int)]
```

```python
# reading validation data

#reading training annotations from excel file
ValDangerYlabel=pd.read_excel("C:\\Users\\molli\\OneDrive\\Desktop\\ValDangerLevel.xlsx")

floodValImg, floodValImgPath = read_data(floodValPath)

# calculate hog features for each image
# yValLabels is 0 for low water logging
# yValLablels is 1 for high water logging 

# hogFeatures(dimension: m x n) contain train features
# y_labels(dimensin: m x 1) contain training labels


hogValFeature=[]
yValLabels=[]
for i in range(len(floodValImg)):
    k=hog_feature(floodValImg[i])
    hogValFeature.append(k)
    currentName= floodValImgPath[i]
    imgno=currentName.split("_")[1]
    imgno=int(imgno.split(".")[0])
    y=ValDangerYlabel.loc[ValDangerYlabel["Image No."]==imgno]["New Level"].item()
    yValLabels.append(y)

# shuffling the validation data
rand_col=np.arange(0,len(yValLabels),dtype=np.int)
np.random.shuffle(rand_col)
yValLabels=np.array(yValLabels)[rand_col.astype(int)]
hogValFeature=np.array(hogValFeature)[rand_col.astype(int)]
```

### Training The Data and Chosing the Best Hyperparameters

```python
def svm_train(hog_features, labels, C_val):
    """ 
    SVM train
    
    Parameters
    ----------
    hog_feature: numpy array hog features.
                 training features, formated as [[hog1], [hog2], ...]
                 
    labels: numpy array labels (1-dim)
            training labels
            
    C_val: float
           val of c parameter for SVM model
        
    Return:
    clf: SVM model weights
    """
    clf = SVC(C=C_val, tol=1e-3, probability = True)
    clf.fit(hog_features, labels)
    return clf
```

```python
def randomforest_train(hog_features, labels,n_estimators ):
    """ 
    Random Forest Train
    
    Parameters
    ----------
    hog_features: numpy array hog features.
                 training features, formated as [[hog1], [hog2], ...]
                 
    labels: numpy array labels (1-dim)
            training labels
            
    n_estimators: int
                  value of no. of trees in random forest
        
    Return:
    classifier: Random Forest model weights
    """
    classifier= RandomForestClassifier(n_estimators=n_estimators)
    classifier.fit(hog_features,labels)
    return classifier
```

```python
#TRAINING RANDOM FOREST MODEL BY CHOSING THE HYPERPARAMETER ON THE VAL SET

# Random Forest classification to predict 1 for water logged image and 0 for non waterlogged image

noOfTrees = [50,100,150,200]
max_accuracy = 0

for trees in noOfTrees:
    # call svm train 
    clf=randomforest_train(hogFeature,y_labels,trees)
    ypred=clf.predict(hogValFeature)
    curr_accuracy = accuracy_score(ypred,yValLabels)
    if curr_accuracy > max_accuracy:
        max_accuracy = curr_accuracy
        best_val = trees
    
print("Best value of hyper parameter is no of trees is ", best_val)
print("The best val accuracy is ", max_accuracy)
```

```python
#TRAINING SVM MODEL BY CHOSING THE HYPERPARAMETER ON THE VAL SET

# SVM classification to predict 1 for water logged image and 0 for non waterlogged image

C = [0.5, 1, 5, 10]
max_accuracy = 0
best_clf= None

for c_val in C:
    # call svm train 
    clf=svm_train(hogFeature,y_labels, c_val)
    ypred=clf.predict(hogValFeature)
    curr_accuracy = accuracy_score(ypred,yValLabels)
    if curr_accuracy > max_accuracy:
        max_accuracy = curr_accuracy
        best_val = c_val
        best_clf=clf
    
print("Best value of hyper parameter is C val is ", best_val)
print("The best val accuracy is ", max_accuracy)
```

```python
#best val accuracy
ypred=best_clf.predict(hogValFeature)
print("best val accuracy ", accuracy_score(ypred,yValLabels))
```
